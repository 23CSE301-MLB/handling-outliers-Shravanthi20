```python
import pandas as pd
import numpy as np
 
Q1 = df['dob_years'].quantile(0.25)
Q3 = df['dob_years'].quantile(0.75)
IQR = Q3 -Q1
lower_bound= Q1 -1.5 * IQR
upper_bound= Q3 + 1.5 * IQR
outliers_iqr= df[(df['dob_years'] < lower_bound) | (df['dob_years'] > upper_bound)]
print(outliers_iqr.shape)
```
